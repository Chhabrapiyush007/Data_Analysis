```python
#Shopping Data Set

# The dataset contains the following columns:

# Administrative: Number of administrative pages visited by the user.

# Administrative_Duration: Total time spent on administrative pages.

# Informational: Number of informational pages visited by the user.

# Informational_Duration: Total time spent on informational pages.

# ProductRelated: Number of product-related pages visited by the user.

# ProductRelated_Duration: Total time spent on product-related pages.

# BounceRates: The proportion of sessions that were a single-page visit (i.e., the session ended without any interaction).

# ExitRates: The proportion of page exits (i.e., the proportion of times a specific page was the last in the session).

# PageValues: The value of a page based on metrics like purchase likelihood.

# SpecialDay: Indicates closeness to a special day (like holidays), where 0.0 means not close and 1.0 means very close.

# Month: The month when the session happened (e.g., 'Feb' for February).

# OperatingSystems: Encoded values representing different operating systems (e.g., 1, 2, 3).

# Browser: Encoded values representing different browsers used by the user.

# Region: Encoded values representing different regions.

# TrafficType: Encoded values representing the source of traffic (e.g., 1, 2, 3).

# VisitorType: Categorical variable indicating whether the visitor is a "Returning_Visitor" or "New_Visitor."

# Weekend: Boolean indicating whether the session was on a weekend (True or False).

# Revenue: Boolean indicating whether the session resulted in a purchase (True or False).
```

```python
#Campaign DATA SET
```

```python
import pandas as pd

# Load the dataset
data = pd.read_csv('/content/campaign - campaign.csv')

# Inspect the first few rows
print(data.head())
print(data.info())
```

```python
# Summary Statistics
```

```python
# Summary statistics for numerical columns
print(data.describe())

# Frequency counts for categorical columns
print(data['Education'].value_counts())
print(data['Marital_Status'].value_counts())
print(data['Country'].value_counts())
```

```python
# Check for Missing Values and Duplicates

# Check for missing values
print(data.isnull().sum())

# Check for duplicate entries
print(data.duplicated().sum())
```

```python
# Feature Engineering
```

```python
# Create a total spending column
data['TotalSpending'] = (data['MntWines'] + data['MntFruits'] +
                         data['MntMeatProducts'] + data['MntFishProducts'] +
                         data['MntSweetProducts'] + data['MntGoldProds'])
```

```python
# Recategorize Marital Status
```

```python
# Map marital statuses to 'In couple' and 'Alone'
data['Marital_Status'] = data['Marital_Status'].replace({
    'Married': 'In couple',
    'Together': 'In couple',
    'Divorced': 'Alone',
    'Single': 'Alone',
    'Absurd': 'Alone',
    'Widow': 'Alone',
    'YOLO': 'Alone'
})
```

```python
# Create Income Brackets
```

```python
# Clean the 'Income' column
data['Income'] = data['Income'].replace({'\$': '', ',': ''}, regex=True).astype(float)

# Calculate the median income
median_income = data['Income'].median()

# Create income brackets
data['IncomeBracket'] = ['Below Median' if x < median_income else 'Above Median' for x in data['Income']]

# Create a column indicating if any campaign was accepted
data['AcceptedAnyCampaign'] = (data['AcceptedCmp1'] | data['AcceptedCmp2'] |
                                data['AcceptedCmp3'] | data['AcceptedCmp4'] |
                                data['AcceptedCmp5'])

# Display the updated DataFrame
print(data.head())
```

```python
# Hypothesis Testing
```

```python
# # Is Income of Customers Dependent on Their Education?

# Hypothesis:

# Null Hypothesis (H0): Income is independent of education.
# Alternative Hypothesis (H1): Income is dependent on education.
# Test: ANOVA (Analysis of Variance) test if Education is a categorical variable and Income is numerical.
```

```python
import statsmodels.api as sm
from statsmodels.formula.api import ols

# ANOVA test
model = ols('Income ~ C(Education)', data=data).fit()
anova_table = sm.stats.anova_lm(model, typ=2)
print(anova_table)
```

```python
# The very low p-value suggests that there is a significant difference in income based on education levels.
# This means that education level does have a statistically significant effect on the income of customers.
```

```python
# # Do Higher Income People Spend More?
# Hypothesis:

# Null Hypothesis (H0): Higher income does not lead to higher total spending.
# Alternative Hypothesis (H1): Higher income leads to higher total spending.
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Scatter plot with regression line
sns.lmplot(x='Income', y='TotalSpending', data=data)
plt.show()

# Calculate Pearson correlation coefficient
correlation = data[['Income', 'TotalSpending']].corr().iloc[0, 1]
print(f'Pearson correlation coefficient: {correlation}')
```

```python
# Based on the high positive correlation, the alternative hypothesis (H1), which states that higher income leads to higher total spending, is supported.
# Higher income individuals tend to spend more in total, reinforcing the idea that income influences spending behavior.
```

```python
# Do Couples Spend More or Less Money on Wine Than People Living Alone?
```

```python
# Hypothesis:

# Null Hypothesis (H0): Couples spend the same amount or less on wine compared to people living alone.
# Alternative Hypothesis (H1): Couples spend more on wine than people living alone.
from scipy.stats import ttest_ind

# Split data by marital status
couples_wine = data[data['Marital_Status'] == 'In couple']['MntWines']
alone_wine = data[data['Marital_Status'] == 'Alone']['MntWines']

# Perform t-test
t_stat, p_val = ttest_ind(couples_wine, alone_wine, equal_var=False)
print(f't-statistic: {t_stat}, p-value: {p_val}')
```

```python
# There is no strong statistical evidence to support the claim that couples spend more on wine compared to people living alone.
# The data does not show a significant difference in wine spending between the two groups.
```

```python
# Are People with Lower Income More Attracted Towards Campaigns?
```

```python
# Hypothesis:

# Null Hypothesis (H0): People with lower income do not accept more campaigns.
# Alternative Hypothesis (H1): People with lower income accept more campaigns.

from scipy.stats import chi2_contingency

# Create contingency table
contingency_table = pd.crosstab(data['IncomeBracket'], data['AcceptedAnyCampaign'])

# Perform Chi-Square test
chi2_stat, p_val, dof, expected = chi2_contingency(contingency_table)
print(f'Chi-square statistic: {chi2_stat}, p-value: {p_val}')
```

```python
# # There is strong statistical evidence to support the claim that people with lower income accept more campaigns.
# The data shows a significant association between lower income and higher campaign acceptance rates.
```
